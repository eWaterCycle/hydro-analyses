### Forecast Ensemble Member Plots

```python
#Create plot for each subbasin showing the ensemble spread.
#Basin Plots
plots = []
curves = []
data_var = list(forecast_data.keys())

for station, observation in grdc_stations:
    curves = [hv.Curve(forecast_data[i].sel(lat = station['grdc_latitude_in_arc_degree'],
                                                      lon = station['grdc_longitude_in_arc_degree'],
                                                      method='nearest'), label=i) for i in data_var]
    
    
    forecast_station_select = forecast_data.sel(lat = station['grdc_latitude_in_arc_degree'],
                                                      lon = station['grdc_longitude_in_arc_degree'],
                                                      method='nearest')
    
    area_minimum = xr.Dataset.to_array(forecast_station_select).min('variable')
    area_maximum = xr.Dataset.to_array(forecast_station_select).max('variable')

    plots.append(hv.Area((forecast_station_select['time'].values, area_maximum.values, area_minimum.values),
                 kdims=[('Date '+ pd.to_datetime(forecast_station_select['time'].values[0]).strftime("%Y")), ('Discharge ['+station['units']+']')],
                 vdims=['max', 'min']).options(title_format=(station['river_name']+' '+station['station_name']+' '+station['country_code'])) *\
                 hv.Overlay.from_values(curves))
```

```python
%%opts Layout [shared_axes=False] Overlay [width=450, height=400 legend_position='right', show_title=True, fontsize={'legend':6, 'title':10, 'xlabel':10, 'ylabel':10, 'ticks':10}] Curve (muted_alpha=0.5 muted_color='black') 

hv.Layout(plots).cols(2)
```

### Forecast Ensemble Mean & GRDC Station Observation Plots

```python
plots = []

for station, observation in grdc_stations:
   
    curves_mean = hv.Curve(forecast_stats['discharge'].sel(lat = station['grdc_latitude_in_arc_degree'],
                                                      lon = station['grdc_longitude_in_arc_degree'],
                                                      method='nearest'), label='Ensemble Mean').options(line_width = 2, color='purple')
    
    forecast_station_select = forecast_data.sel(lat = station['grdc_latitude_in_arc_degree'],
                                                      lon = station['grdc_longitude_in_arc_degree'],
                                                      method='nearest')
    
    area_minimum = xr.Dataset.to_array(forecast_station_select).min('variable')
    area_maximum = xr.Dataset.to_array(forecast_station_select).max('variable')
    
    observation.index = observation.index + pd.offsets.DateOffset(years=10)
    grdc_observation = hv.Curve(observation, label='GRDC observation 2008 data').options(line_width = 2, color='red')
    
    plots.append(hv.Area((forecast_station_select['time'].values, area_maximum.values, area_minimum.values),
                 kdims=[('Date '+ pd.to_datetime(forecast_station_select['time'].values[0]).strftime("%Y")), ('Discharge ['+station['units']+']')],
                 vdims=['_max', '_min']).options(title_format=(station['river_name']+' '+station['station_name']+' '+station['country_code'])) *\
                 hv.Overlay.from_values(curves_mean) *\
                 hv.Overlay.from_values(grdc_observation))
    

                 
```

### Forecast Ensemble Mean & GRDC Station Observation Statistics

```python
for station, observation in grdc_stations:
   
    forecast_mean_station_select = forecast_stats['discharge'].sel(lat = station['grdc_latitude_in_arc_degree'],
                                                      lon = station['grdc_longitude_in_arc_degree'],
                                                      method='nearest')

    sim_df = xr.DataArray.to_dataframe(forecast_mean_station_select)
    sim_df = sim_df.drop(columns=['lon', 'lat'])
    
    obs_df = observation
    
    
    merged_df = hydrod.merge_data(sim_df=sim_df, obs_df=obs_df, column_names=('Mean of Ensemble', 'GRDC observation'))
    
    hydrov.plot(merged_data_df=merged_df,
        title=('Hydrograph of ' + station['river_name']+' '+station['station_name']+' '+station['country_code']),
        linestyles=['r-', 'k-'],
        legend=('Forecast ensemble mean', 'GRDC observation 2008 data'),
        labels=['Date', ('Discharge ['+station['units']+']')],
        metrics=['ME', 'NSE', 'SA'],
        grid=True)


    hydrov.hist(merged_data_df=merged_df,
        num_bins=20,
        title=('Histogram of Streamflows '+ station['river_name']+' '+station['station_name']+' '+station['country_code']),
        legend=('Forecast ensemble mean', 'GRDC observation 2008 data'),
        labels=['Bins', 'Frequency'],
        grid=True)
    
    hydrov.scatter(merged_data_df=merged_df,
        grid=True,
        title=('Scatter Plot (Normal Scale) of ' + station['river_name']+' '+station['station_name']+' '+station['country_code']),
        labels=['GRDC observation 2008 data', 'Forecast ensemble mean'],
        line45=True,
        metrics=['ME', 'KGE (2012)'],
        best_fit=True)


    hydrov.qqplot(merged_data_df=merged_df,
        title=('Quantile-Quantile Plot of ' + station['river_name']+' '+station['station_name']+' '+station['country_code']),
        xlabel='GRDC observation 2008 data Quantiles',
        ylabel='Forecast ensemble mean Quantiles',
        legend=True,
        )
    plt.show()
```
