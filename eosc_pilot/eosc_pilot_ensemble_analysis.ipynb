```python
# stack the observations and the ensemble results ready for plotting a rank histogram
for station in stations:
    station.stack = np.stack(station.array, axis=0) # ensemble stack
ens_stack = np.stack((station.stack for station in stations), axis=2)
obs_stack = np.stack((station.obs_data for station in stations), axis=1)
```

```python
# see https://github.com/oliverangelil/rankhistogram/blob/master/README.md
no_time_steps = len(obs_stack)
no_stations = len(stations)

obs = np.array(obs_stack)
ensemble = np.array(ens_stack)

#mask = np.random.randint(0, 2, (no_time_steps, no_stations)) #masked where 0/false.
mask = np.ones((no_time_steps, no_stations)) # no mask, masked where 0/false

# feed into rankz function
result = rankz(obs, ensemble, mask)

# plot histogram
plt.bar(range(1,ensemble.shape[0]+2), result[0])
plt.xlabel("Rank of Observation")

# view histogram
plt.show()
```
